```python
import matplotlib.pyplot as plt
import pandas
import numpy as np
```

```python
# load data
file = '../Data/csv/brazil_legal_amazon_deforestation.csv'

data = pandas.read_csv(file)

'''
File contains 3 records of annual deforestation rates 
for the Brazilian Legal Amazon. 

The ESA and MODIS change is calculated by subtracting total forest areas
of 2 subsequent years (e.g. 2002-2001, which is assigned to 2001).

The PRODES data (available: http://terrabrasilis.dpi.inpe.br/app/dashboard/deforestation/biomes/legal_amazon/rates)
is based on analysis of imagery from multiple satellites, including Landsat. 
More information is available here: http://www.obt.inpe.br/OBT/assuntos/programas/amazonia/prodes
'''
```

```python
# plot - code adapted from November Sense training
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(7,4))
axes.plot(data['Year'], -data['ESA_CCI'], color='royalblue', linewidth=2.0, linestyle='--', label = 'ESA')
axes.plot(data['Year'], -data['MOD12C1'], color='r', linewidth=2.0, linestyle='--', label = 'MODIS')
axes.plot(data['Year'], -data['PRODES'], color='black', linewidth=1.5, linestyle='-', label = 'PRODES');

axes.set_ylabel('Area deforested (km$^2$)')
axes.legend()

x_ticks = np.linspace(2001, 2019, 10)
#x_labels = [r'- $\pi$', r'- $\pi$/2', '0', r'$\pi$/2', r'$\pi$']

axes.set_xticks(x_ticks)
#axes.set_xticklabels(x_labels);

plt.savefig('Figures/braz_leg_amazon_comparison_full.png', dpi = 300)
```

```python
# plot - code adapted from November Sense training
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(6,4))
axes.plot(data['Year'], -data['ESA_CCI']/1000, color='royalblue', linewidth=2.0, linestyle='--', label = 'ESA')
axes.plot(data['Year'], -data['MOD12C1']/1000, color='r', linewidth=2.0, linestyle='--', label = 'MODIS')
axes.plot(data['Year'], -data['PRODES']/1000, color='black', linewidth=1.5, linestyle='-', label = 'PRODES');

axes.set_ylim(-2, 40)
axes.set_ylabel('Area deforested (1000 km$^2$)')

axes.legend()

x_ticks = np.linspace(2001, 2019, 10)
y_ticks = np.linspace(0, 40, 5)
#x_labels = [r'- $\pi$', r'- $\pi$/2', '0', r'$\pi$/2', r'$\pi$']

axes.set_xticks(x_ticks)
axes.set_yticks(y_ticks)
#axes.set_xticklabels(x_labels);

plt.savefig('Figures/braz_leg_amazon_comparison_closeup.png', dpi = 300)
```
